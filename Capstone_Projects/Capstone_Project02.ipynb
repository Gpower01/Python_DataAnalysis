- **Step 2: I will create a list with the ticker symbols in alphabetical order and I will call this tickers**
- Use pd.concat() to concatenate the bank DataFrames together into a single DataFrame called bank_stocks. Setting the keys argument equal to tickers list. Also paying attention to what axis to concatenate on.

```python
# To concatenate into a single column, I will pass axis=1 and the keys argument to tickers
bank_stocks = pd.concat([BAC, C, GS, JPM, MS, WFC],axis=1,keys=tickers)
```

- Now I will create a new empty DataFrame called **returns**. This dataframe will contain the returns for each bank's stock. Returns are typically defined by:

$$r_t = \frac{p_t - p_{t-1}}{p_{t-1}} = \frac{p_t}{p_{t-1}} - 1$$


```python
returns = pd.DataFrame()
```

- I can use pandas **pct_change()** method on the Close column to create a column representing this return value. Create a for loop that goes and for each Bank Stock Ticker creates this returns column and set's it as a column in the returns DataFrame.

```python
# Create a 'for loop'
for tick in tickers:
    returns[tick+ 'Return'] = bank_stocks[tick]['Close'].pct_change()
```
